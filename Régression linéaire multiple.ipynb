# Implémentation: modèle prédictif

```python
pd_dataset["MEDV"] = dataset.target
```

```python
pd_dataset.head()
```

```python
def lin_reg(b,W,X):
    y = 0
    for xi,wi in zip(X,W):
        y= y + xi*wi
    y += b 
    return y
```

```python
medv = pd_dataset["MEDV"].values
lstat = pd_dataset["LSTAT"].values
rm = pd_dataset["RM"].values
```

```python
print("MEDV", medv[:10])
print("LSTAT", lstat[:10])
print("RM", rm[:10])
```

```python
b = 0
W = [1,1]
Y = []

for l,r in zip(lstat, rm):
    print(l,r)
```

```python
b = 0
W = [1,1]
Y = []

for l,r in zip(lstat, rm):
    X=[l,r]
    y=lin_reg(b,W,X)
    Y.append(y)
    
print(Y[:10])
print(medv[:10])
```

```python
plt.plot(Y, label="Prediction")
plt.plot(medv, label="prix")
plt.legend
plt.show()
```

```python
N=100
plt.plot(Y[:N], label="Prediction")
plt.plot(medv[:N], label="prix")
plt.legend
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(medv,Y)
mse
```

# Implémentation: scikit-learn

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = load_boston()
pd_dataset = pd.DataFrame(dataset.data, columns=dataset.feature_names)
pd_dataset['MEDV'] = dataset.target
pd_dataset.head()
```

```python
X = pd_dataset[['RM', 'LSTAT']]
```

```python
X.head()
```

```python
X = pd_dataset[['RM', 'LSTAT']].values
Y = pd_dataset['MEDV'].values
```

```python
modele = linear_model.LinearRegression()
modele.fit(X, Y)
```

```python
print("Coefs", modele.coef_)
print("Biais", modele.intercept_)
```

```python
Y_pred = modele.predict(X)

print("Prix", Y[:10])
print("Prédictions", Y_pred[:10])
```

```python
N = 100
plt.plot(Y_pred[:N], label='prédictions')
plt.plot(Y[:N], label='Prix')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Y, Y_pred)
mse
```
