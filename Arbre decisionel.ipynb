```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
# Importation des données
df = pd.read_csv('https:/TUTORIALS/main/cancer.csv')
# Affichage de df
df.head()
```

```python
df.info()
```

```python
# Diagramme à barre de la variable 'diagnosis'
sns.countplot('diagnosis', data = df).set_title("Distribution de la variable 'diagnosis'");
```

```python
# Suppression des variables 'id' et 'Unnamed: 32'
df.drop(columns = ['id', 'Unnamed: 32'], inplace = True)
```

```python
# Variable cible
y = df['diagnosis']
```

```python
# variables indépendantes
X = df.drop('diagnosis', axis = 1)
```

```python
# Données d'entraînement (80%) et données de test (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1111)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Création d'un modèle d'arbre de décision
tree_model = DecisionTreeClassifier()
```

```python
# Entraînement du modèle
tree_model.fit(X_train, y_train)
```

```python
# Précision globale sur les données d'entraînement
tree_model.score(X_train, y_train)
```

```python
# Précision globale sur les données de test
tree_model.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 4))
ax = fig.add_axes([0, 0, 0.8, 1], frameon=False, xticks=[], yticks=[])
ax.set_title("Exemple d'arbre de décision : classification des animaux", size=24)

def text(ax, x, y, t, size=20, **kwargs):
    ax.text(x, y, t,
            ha='center', va='center', size=size,
            bbox=dict(boxstyle='round', ec='k', fc='w'), **kwargs)

text(ax, 0.5, 0.9, "Quelle est la taille \n de l'animal ?", 20)
text(ax, 0.3, 0.6, "Est-ce que l'animal \n a des cornes ?", 18)
text(ax, 0.7, 0.6, "Est-ce que l'animal \n a deux pattes ?", 18)
text(ax, 0.12, 0.3, "Les cornes \nplus de 10 cm?", 14)
text(ax, 0.38, 0.3, "Est-ce que l'animal \n porte un collier?", 14)
text(ax, 0.62, 0.3, "Est-ce que l'animal \n a des ailes?", 14)
text(ax, 0.88, 0.3, "Est-ce que l'animal \n a une queue ?", 14)

text(ax, 0.4, 0.75, "> 1m", 12, alpha=0.4)
text(ax, 0.6, 0.75, "< 1m", 12, alpha=0.4)

text(ax, 0.21, 0.45, "Oui", 12, alpha=0.4)
text(ax, 0.34, 0.45, "non", 12, alpha=0.4)

text(ax, 0.66, 0.45, "Oui", 12, alpha=0.4)
text(ax, 0.79, 0.45, "non", 12, alpha=0.4)

ax.plot([0.3, 0.5, 0.7], [0.6, 0.9, 0.6], '-k')
ax.plot([0.12, 0.3, 0.38], [0.3, 0.6, 0.3], '-k')
ax.plot([0.62, 0.7, 0.88], [0.3, 0.6, 0.3], '-k')
ax.plot([0.0, 0.12, 0.20], [0.0, 0.3, 0.0], '--k')
ax.plot([0.28, 0.38, 0.48], [0.0, 0.3, 0.0], '--k')
ax.plot([0.52, 0.62, 0.72], [0.0, 0.3, 0.0], '--k')
ax.plot([0.8, 0.88, 1.0], [0.0, 0.3, 0.0], '--k')
ax.axis([0, 1, 0, 1])
```
