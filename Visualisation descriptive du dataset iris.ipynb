# Visualisation de dataset IRIS

```python
# On importe les modules necessaires pour la visualisation 

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv("iris.csv")
    # On importe notre base de donn√©es iris dans le dataframe
```

```python
print(dataset.shape)
    # On a 150 lignes et 5 colonnes
```

```python
dataset["variety"].value_counts()
    # Le dataset contient 50 especes pour chaque categorie de fleurs
```

```python
dataset.describe()
```

```python
plt.plot(dataset["sepal.length"], "o")
plt.show()
```

```python
plt.plot(dataset["sepal.width"], "o")
plt.show()
```

```python
plt.plot(dataset["sepal.length"], dataset["sepal.width"], "o")
plt.show()

    # On peut remarquer une certaine tendance de donn√©es
```
